```python
import time
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(n):
    x = 1
    for i in range(n):
        for j in range(n):
            x += 1
    return x


n_values = list(range(1, 101))  # Test for n from 1 to 100
times = []

for n in n_values:
    start_time = time.time()
    f(n)
    end_time = time.time()
    times.append(end_time - start_time)

# Polynomial fit (n^2 expected)
coefficients = np.polyfit(n_values, times, 2)
fitted_curve = np.polyval(coefficients, n_values)

# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(n_values, times, 'o', label="Measured Times")
plt.plot(n_values, fitted_curve, '-', label=f"Fitted Curve: {coefficients[0]:.2e}n^2 + {coefficients[1]:.2e}n + {coefficients[2]:.2e}")
plt.xlabel("n (Input Size)", fontsize=14)
plt.ylabel("Time (Seconds)", fontsize=14)
plt.title("Execution Time vs. Input Size", fontsize=16)
plt.legend(fontsize=12)
plt.grid()
plt.show()
```
